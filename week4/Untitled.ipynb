```python
import pandas as pd
import numpy as np

# Sample DataFrame (replace this with your actual data)
df = pd.DataFrame({
    'Age': [25, np.nan, 30, 28, np.nan],
    'Gender': ['Male', 'Female', np.nan, 'Male', 'Female'],
    'Salary': [50000, 54000, 60000, np.nan, 58000]
})

# üîç 1. Check for missing values
print("Missing values in each column:\n", df.isnull().sum())

# üîÅ 2. Replace missing values
# Numerical columns ‚Äì replace with mean
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(df[col].mean(), inplace=True)

# Categorical columns ‚Äì replace with mode
for col in df.select_dtypes(include=['object']):
    df[col].fillna(df[col].mode()[0], inplace=True)

# ‚úÖ Final cleaned DataFrame
print("\nCleaned DataFrame:\n", df)
```
